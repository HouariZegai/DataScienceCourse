# Beautiful Soup
### for get data from any thing (website, ...)

```python
# import libraries
import bs4 as bs
import urllib.request
from urllib.request import urlopen

url = urlopen('http://stackoverflow.com').read()
```

```python
soup = bs.BeautifulSoup(url, 'lxml')
soup
```

```python
print("Title: ", soup.title,
      "\nText: ", soup.title.text,
     "\nname: ", soup.title.name)
```

```python
parag = soup.find_all('p') # get all p tag
for p in parag:
    print(p.text))
```

```python
# Get all links
for url in soup.find_all('a'):
    if url.get('href') == '#':
        continue
    print(url.get('href'))
```

```python
body = soup.body
 # from body tag get all div tags with class 'related-links'
for div in body.find_all('div', class_ = "related-links"):
    print(div)
```

```python
# get all src(url) of img
for img in body.find_all('img'):
    print(img.get('src'))
```
